Student Performance Tracker

- store multiple students with name, marks, attendance in dictionary.

Use functions to

1. Add a student
2. Update marks / attendance
3. Calculate Grade
4. Display Report

- Handle exceptions of user enters invalid data.



```python
# Student Performance Tracker

students = {}

def add_student(name, marks, attendance):
  try:
    students[name] = {"marks": int(marks),
                     "attendance" : int(attendance)
                     }
    print(f"{name} added successfully")

  except ValueError:
    print("Invalid Input, Marks and attendance must be numbers.")


def update_marks(name, marks):
  try:
    if name in students:
      students[name]["marks"] = int(marks)
      print(f"Marks successfully updated")
    else:
      print(f"{name} not found")

  except ValueError:
    print("Enter valid marks in numbers")

def update_attendance(name, attendance):
  try:
    if name in students:
      students[name]["attendance"] = int(attendance)
      print(f"Attendance successfully updated")
    else:
      print(f"{name} not found")

  except ValueError:
    print("Attendance should be a number")

def calculate_grade(marks):
  if marks >= 90:
    return 'A'
  elif marks >= 70:
    return 'B'
  elif marks >= 50:
    return 'C'
  else:
    return 'FAIL'

def display_report():
  print("/n Student Report Card")
  print("-" * 40)

  for name, details in students.items():
    grade = calculate_grade(details["marks"])
    status = "Pass" if grade != "FAIL" else "Fail"

    print(f"{name} | Marks: {details['marks']} | Attendance: {details['attendance']} | Grade: {grade} | Status: {status}")
    print("-" * 40)

# Main program
while True:
  print("-" * 40)
  print("\nStudent Performance Tracker")
  print("-" * 40)
  print("1. Add Student")
  print("2. Update Marks")
  print("3. Update Attendance")
  print("4. Display Report")
  print("5. Exit")

  choice = input("Enter your choice: ")

  if choice == '1':
    name = input("Enter student name: ")
    marks = input("Enter marks: ")
    attendance = input("Enter attendance: ")
    add_student(name, marks, attendance)

  elif choice == '2':
    name = input("Enter student name: ")
    marks = input("Enter new marks: ")
    update_marks(name, marks)

  elif choice == '3':
    name = input("Enter student name: ")
    attendance = input("Enter new attendance: ")
    update_attendance(name, attendance)

  elif choice == '4':
    display_report()

  elif choice == '5':
    print("Exiting the program.")
    break

  else:
    print("Invalid choice. Please try again.")


```

Todo List with Priority (Task Manager)

Build a Task Manager that can

*   Add a task with priority(High, medium, low)


*   View all task sorted with priority
*   Mark a task as done


*   Delete a task
*   Handle Invalid input using exceptions






```python
# Task Manager with priority

tasks = []

def add_task():
  try:
    task = input("Enter task: ")
    priority = input("Enter priority (High, Medium, Low): ").capitalize()


    if priority.lower() not in ['high', 'medium', 'low']:
      raise ValueError

    tasks.append({"task": task, "priority": priority, "done" : False})
    tasks.sort(key=lambda x: ["High", "Medium", "Low"].index(x["priority"]))
    print(f"Task {task} added successfully with {priority} priority")
  except ValueError as ve:
    print("Error!",ve)

def view_tasks():
  if not tasks:
    print("No tasks added yet")
    return
  print("\n Tasks List:")
  sorted_tasks = sorted(tasks, key=lambda x:["High", "Medium", "Low"].index(x["priority"]))
  for i, task in enumerate(tasks, start=1):
    status = "Done" if task["done"] else "Not Done"
    print(f"{i}. Task: {task['task']} | Priority: {task['priority']} | Status: {status}")

def mark_done():
  try:
    view_tasks()
    task_number = int(input("Enter the task number to mark as done: "))
    if task_number < 1 or task_number > len(tasks):
      raise IndexError ("Task number out of range")
    tasks[task_number - 1]["done"] = True
    print("Task marked as done")
  except (ValueError, IndexError) as e:
    print("Error!",e)

def delete_task():
  try:
    view_tasks()
    task_number = int(input("Enter the task number to delete: "))
    if task_number < 1 or task_number > len(tasks):
      raise IndexError ("Task number out of range")
    removed = tasks.pop(task_number - 1)
    print(f"Task '{removed['task']}' deleted successfully")
  except (ValueError, IndexError) as e:
    print("Error!",e)


# main program

while True:
  print("\n--- Task Manager menu ---")
  print("1. Add Task")
  print("2. View Tasks")
  print("3. Mark Task as Done")
  print("4. Delete Task")
  print("5. Quit")

  choice = input("Enter your choice: ")

  if choice == '1':
    add_task()
  elif choice == '2':
    view_tasks()
  elif choice == '3':
    mark_done()
  elif choice == '4':
    delete_task()
  elif choice == '5':
    print("Exiting task manager. See you soon!")
    break
  else:
    print("Invalid choice. Please try again.")
```
